```python
df.loc[:, 'expense_to_income_ratio'] = df['monthly_expenses_usd'] / df['monthly_income_usd']
df.loc[:, 'savings_to_expense_ratio'] = df['savings_usd'] / df['monthly_expenses_usd']
df.loc[:, 'age_group'] = pd.cut(df['age'], bins=[18,30,45,60,100], labels=['Young','Adult','Mid-age','Senior'])
df
```

```python
cat_cols = ['gender','education_level','employment_status','job_title',
            'has_loan','loan_type','region','age_group']
df[cat_cols] = df[cat_cols].astype('category')
```

## BUSINESS OBJECTIVES

```python
#1.How can we find customers who are most likely to default on loans?
default_customers = df[
    (df['debt_to_income_ratio'] > 0.5) & 
    (df['credit_score'] < 600) & 
    (df['monthly_emi_usd'] > 0.3 * df['monthly_income_usd'])
]
counts = default_customers['loan_type'].value_counts()
counts.plot(kind='bar', color='skyblue')
plt.title('High-Default-Risk Customers by Loan Type')
plt.xlabel('Loan Type')
plt.ylabel('Count of Risky Customers')
plt.show()
```

### Analysis
If certain loan types (like personal or credit card loans) dominate the graph, it indicates these are riskier categories.

Financial institutions should tighten eligibility checks for these loans or offer better repayment support.

This insight helps in proactive risk management and minimizing defaults.

```python
#2.Which types of loans are most popular among different age groups, jobs, and regions?
loan_age_counts = df[df['has_loan']=='Yes'].groupby(['age_group','loan_type']).size().unstack(fill_value=0)
loan_age_counts.plot(kind='bar', figsize=(8,6))
plt.title('Loan Type Popularity by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Loans')
plt.show()
```

### Analysis
Younger groups may lean toward education or personal loans.

Middle-aged groups might prefer home or vehicle loans.

Older groups may have fewer loans but possibly more mortgage-related.

This pattern helps banks design age-specific loan products.

```python
#3.How can we suggest better savings and investment plans to customers?
df['savings_to_income'] = df['savings_usd'] / df['monthly_income_usd']
plt.scatter(df['monthly_income_usd'], df['savings_to_income'], alpha=0.5)
plt.title('Savings Rate vs Income')
plt.xlabel('Monthly Income (USD)')
plt.ylabel('Savings-to-Income Ratio')
plt.show()
```

### Analysis
A cluster of customers with low savings ratios across income levels suggests poor financial planning.

Higher-income customers should ideally save more, but if the plot shows the opposite, it means spending habits dominate.

This helps in targeting personalized savings plans or investment products for different income groups.

```python
#4.Which regions have the highest demand for loans or the most financial stress?
loan_count_region = df[df['has_loan']=='Yes']['region'].value_counts()
loan_count_region.plot(kind='bar', color='salmon')
plt.title('Number of Loan Customers by Region')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.show()

avg_dti_region = df.groupby('region')['debt_to_income_ratio'].mean()
avg_dti_region.plot(kind='bar', color='olive')
plt.title('Average Debt-to-Income by Region')
plt.xlabel('Region')
plt.ylabel('Average DTI')
plt.show()
```

### Analysis:

Regions with high DTI are more financially stressed.

Even if the loan count is low, a high DTI indicates repayment struggles.

This data guides regional financial literacy campaigns or localized credit policies.

```python
#5.Who are the customers with low savings or poor credit, and how can we support them?
low_savings = df[df['savings_to_income_ratio'] < 0.1]
poor_credit = df[df['credit_score'] < 600]
sns.histplot(poor_credit['credit_score'], bins=20)
plt.title("Credit Score Distribution (Poor Credit)")
plt.show()
```

### Analysis:

If most poor-credit customers cluster between 400–500, these are high-risk individuals.

Lenders must consider financial rehabilitation programs, stricter loan approvals, or secured loan products.

This also shows the size of the at-risk customer base.
